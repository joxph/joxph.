```python
#libray imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dataset insertion
data = pd.read_csv('/breastcancer.csv')
data.head()
```

```python
data.info()
```

```python
data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diagnosis']= le.fit_transform(data['diagnosis'])
```

```python
data.info()
```

```python
data.head()
```

```python
#droping colums
X = data.drop('diagnosis', axis=1)
y = data.diagnosis
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = models.Sequential()

model.add(layers.Dense(8, activation='relu', input_shape=(30,)))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=5)
```

```python

model1 = models.Sequential()

model1.add(layers.Dense(64, activation='relu', input_shape=(30,)))

model1.add(layers.Dense(32, activation='relu'))
model1.add(layers.Dense(64, activation='sigmoid'))
model1.add(layers.Dense(64, activation='relu'))
model1.add(layers.Dense(32, activation='relu'))
model1.add(layers.Dense(8, activation='relu'))

model1.add(layers.Dense(1, activation='sigmoid'))
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model1.fit(X_train, y_train, epochs=20)
```

```python
print("accuracy, loss of first model", model.evaluate(X_test, y_test))
print("accuracy, loss of second model", model1.evaluate(X_test, y_test))
```

```python
model2 = models.Sequential()

model2.add(layers.Dense(64, activation='relu', input_shape=(30,)))

model2.add(layers.Dense(8, activation='relu'))
model2.add(layers.Dense(8, activation='relu'))
model2.add(layers.Dense(8, activation='relu'))

model2.add(layers.Dense(1, activation='sigmoid'))
```

```python
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model2.fit(X_train, y_train, epochs=20)
```

```python
print("accuracy, loss of first model", model.evaluate(X_test, y_test))
print("accuracy, loss of second model", model1.evaluate(X_test, y_test))
print("accuracy, loss of third model", model2.evaluate(X_test, y_test))
```
